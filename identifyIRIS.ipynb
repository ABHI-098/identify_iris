```python
import pandas as pd
from sklearn.datasets import load_iris
irs=load_iris()
```

```python
irs.feature_names
```

```python
df=pd.DataFrame(irs.data,columns=irs.feature_names)
```

```python
df['target']=irs.target
```

```python
df['flowerName']=df.target.apply(lambda i:irs.target_names[i])
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X=df.drop(['target','flowerName'],axis='columns')
Y=df.target
Xtrain,Xtest,Ytrain,Ytest=tts(X,Y,test_size=0.21)
```

```python
len(Xtrain)
```

```python
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(Xtrain,Ytrain)
```

```python
model.score(Xtest,Ytest)
```

```python
model.predict([[4.6,4.4,1.2,0.4]])
```
